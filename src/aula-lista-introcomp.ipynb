## Dinâmica da Caixa

```python
import random
import time

# 1. Boas-vindas e Configuração Inicial
# ----------------------------------------
print("--- 🎲 Bem-vindo(a) à Roleta (Versão com Listas)! ---")
time.sleep(1)

# Esta é a LISTA de valores que serão sorteados para formar o saldo do jogador.
valores_pontos = [5, 10, 25, 50, 100, 500]

num_jogadores = int(input("\nQuantos jogadores vão participar? "))

# 2. Coleta de Apostas usando Listas
# ---------------------------------------------
# Vamos usar várias LISTAS para guardar as informações. Cada lista guarda
# um tipo de dado, e a POSIÇÃO (índice) conecta as informações do mesmo jogador.

# Exemplo para 2 jogadores (Ana no índice 0, Beto no índice 1):
# nomes_jogadores =   ["Ana", "Beto"]
# numeros_apostados = [10,    25]
# saldos_iniciais =   [150,   85]

nomes_jogadores = []
numeros_apostados = []
saldos_iniciais = []
total_apostado = 0

print("\n--- Hora de fazer as apostas! ---")
# O `for` vai repetir o bloco de código para cada jogador.
for i in range(num_jogadores):
    print(f"\n--- Vez do Jogador {i+1} ---")

    nome = input(f"Digite o nome do jogador {i+1}: ")

    # Calcula o saldo inicial sorteando 3 fichas e somando.
    saldo = 0
    for _ in range(3):
        ficha_sorteada = random.choice(valores_pontos)
        saldo += ficha_sorteada

    print(f"💰 {nome}, você sorteou 3 fichas e seu saldo é de {saldo} pontos.")

    numero = int(input(f"Em qual número você quer apostar (de 0 a 36)? "))

    # O jogador aposta todo o seu saldo (All-in).
    print(f"Você apostou todos os seus {saldo} pontos no número {numero}.")

    # Adiciona as informações do jogador nas nossas listas.
    # O método `.append()` adiciona um item ao FINAL da lista.
    # A ordem é importante para manter os dados alinhados!
    nomes_jogadores.append(nome)
    numeros_apostados.append(numero)
    saldos_iniciais.append(saldo)

    # Soma a aposta ao total. Como é 'all-in', a aposta é o saldo.
    total_apostado += saldo

# 3. Girando a Roleta
# -------------------
print("\n----------------------------------------------------")
print("\nTodas as apostas foram feitas!")
print(f"O valor total na mesa é de {total_apostado} pontos.")
print("\n🎰 A roleta está girando...")
time.sleep(3) # Pausa maior para mais suspense.

numero_vencedor = random.randint(0, 36)
print(f"\n🎉 O número sorteado é: {numero_vencedor}!")
time.sleep(2)

# 4. Verificando quem ganhou usando os ÍNDICES da lista
# ------------------------------------------------------
print("\n--- 🏆 Resultado Final 🏆 ---")

# Vamos criar uma LISTA para guardar os ÍNDICES dos jogadores que ganharam.
indices_vencedores = []
# `range(num_jogadores)` vai gerar os índices: 0, 1, 2...
for i in range(num_jogadores):
    # Verificamos se o número na lista `numeros_apostados` na posição `i` é o vencedor.
    if numeros_apostados[i] == numero_vencedor:
        # Se for, guardamos o ÍNDICE `i` na nossa lista de vencedores.
        indices_vencedores.append(i)

# 5. Calculando e Mostrando os Resultados
# -----------------------------------------
# `len()` retorna o tamanho da lista.
if len(indices_vencedores) > 0:
    premio_por_vencedor = total_apostado / len(indices_vencedores)
    print(f"\nTemos {len(indices_vencedores)} vencedor(es)! O prêmio de {total_apostado} pontos será dividido.")

    # Passamos pela lista de ÍNDICES dos vencedores.
    for i in indices_vencedores:
        # Usamos o índice `i` para pegar o nome do vencedor na lista `nomes_jogadores`.
        print(f"-> Parabéns, {nomes_jogadores[i]}! Você ganhou {premio_por_vencedor:.2f} pontos.")
else:
    print("\n😔 Ninguém acertou o número! A casa (o computador) ganha.")

# 6. Balanço Final de Pontos
# ----------------------------
print("\n--- 📊 Balanço Final dos Jogadores ---")
saldos_finais = [] # Criamos uma nova lista para guardar os saldos finais
for i in range(num_jogadores):
    # Pegamos as informações do jogador na posição `i` de cada lista.
    nome_jogador = nomes_jogadores[i]
    saldo_inicial_jogador = saldos_iniciais[i]

    # Como a aposta foi 'all-in', o saldo inicial vira 0, a não ser que ganhe.
    saldo_final = 0

    # Se o índice `i` estiver na lista de vencedores, o jogador recebe o prêmio.
    if i in indices_vencedores:
        saldo_final = premio_por_vencedor

    saldos_finais.append(saldo_final) # Adicionamos o saldo final à lista

    print(f"\nJogador: {nome_jogador}")
    print(f"  - Saldo Inicial: {saldo_inicial_jogador} pontos")
    print(f"  - Saldo Final: {saldo_final} pontos")

# 7. Exibindo as Listas Finais para Demonstração
# -----------------------------------------------
print("\n----------------------------------------------------")
print("Conteúdo final das listas usadas no jogo:")
print(f"  -> nomes_jogadores:   {nomes_jogadores}")
print(f"  -> numeros_apostados: {numeros_apostados}")
print(f"  -> saldos_iniciais:   {saldos_iniciais}")
print(f"  -> saldos_finais:     {saldos_finais}")
print(f"  -> indices_vencedores:  {indices_vencedores}")

print("\n--- Fim do Jogo ---")
```

## 2. Introdução a Listas

Uma **lista** em Python é uma coleção ordenada e mutável de itens. Isso significa que os itens têm uma ordem definida (e essa ordem não mudará a menos que você a altere) e você pode mudar, adicionar ou remover itens depois que a lista foi criada.

Os itens em uma lista, também chamados de **elementos**, podem ser de qualquer tipo de dado, e uma única lista pode até conter itens de tipos diferentes (embora seja mais comum ter listas com itens do mesmo tipo).

Cada item em uma lista possui um **índice**, que é um número indicando sua posição. A contagem dos índices começa em `0`. Ou seja, o primeiro item tem índice `0`, o segundo item tem índice `1`, e assim por diante.

**Sintaxe para criar uma lista:**


> < nome da lista > = [ elemento1, elemento2, elemento3, ...]



```python
valores_pontos = [5, 10, 25]
print("Valores dos pontos:", valores_pontos)
print("-" * 20)
```

Para acessar um item específico da lista, usamos seu índice entre colchetes:

```python
print("Acessando Itens da Lista:\n")
primeiro_valor = valores_pontos[0]
terceiro_valor = valores_pontos[2]
print("Primeiro valor:", primeiro_valor)
print("Terceiro valor:", terceiro_valor)
print("-" * 20)
```

Como as listas são mutáveis, podemos alterar um item existente:

```python
print(f"Lista antes da alteração: {valores_pontos}\n")
valores_pontos[2] = 50
print(f"Lista depois da alteração: {valores_pontos}")
```

###3. Adicionar e Remover da lista

Listas são dinâmicas, o que significa que podemos adicionar e remover itens delas facilmente.

### 3.1. Adicionando itens

#### `append()`: Adiciona um item ao final da lista.

```python
nomes_jogadores = ["Luiz", "Artur"]
print(f"Lista inicial: {nomes_jogadores}")

print("\n## Adicionando Itens à Lista\n")
print("Valores antes de adicionar:", nomes_jogadores)
nomes_jogadores.append("Rafael")
nomes_jogadores.append("Guilherme")
print("Valores depois de adicionar novos nomes:", nomes_jogadores)
print("-" * 20)
```

#### `insert()`: Adiciona um item em uma posição específica.
A sintaxe é `lista.insert(indice, item)`.

```python
nomes_jogadores = ["Luiz", "Artur"]
print(f"Lista inicial: {nomes_jogadores}")

nomes_jogadores.insert(1, "Guilherme") # Insere "Guilherme" no índice 1
print(f"Depois de insert(1, 'Guilherme'): {nomes_jogadores}")

nomes_jogadores.insert(2, "Rafael") # Insere "Artur" no início
print(f"Depois de insert(2, 'Artur'): {nomes_jogadores}")
```

### 3.1. Removendo itens

#### `pop()`: Remove e retorna o item de uma posição específica. Se nenhum índice for fornecido, remove e retorna o último item.

```python
valores_pontos = ["5", "10", "25", "50", "100", "500"]

print("## Removendo Itens da Lista")
print(f"\nValores antes de remover: {valores_pontos}\n")
# Removendo o valor 10, 25, 500
valores_pontos.remove("10")
valores_pontos.remove("25")
valores_pontos.remove("500")

print(f"Valores depois de remover o 10, 25 e 500: {valores_pontos}\n")

valores_pontos.pop(2)
print(f"Valores depois de remover o item no índice 2: {valores_pontos}\n")
print("-" * 20)
```

#### `remove()`: Remove a primeira ocorrência de um item com um valor específico.
Se o item não existir na lista, um erro (`ValueError`) será gerado.

```python
convidados = ["Alice", "Bob", "Charlie", "David", "Bob"]
print(f"Convidados iniciais: {convidados}")

convidados.remove("Bob") # Remove a primeira ocorrência de "Bob"
print(f"Depois de remove('Bob'): {convidados}")

convidados.remove("David")
print(f"Depois de remove('David'): {convidados}")

# Tentando remover um item que não existe (isso geraria um erro, então vamos verificar antes)
item_a_remover = "Eve"
if item_a_remover in convidados:
    convidados.remove(item_a_remover)
    print(f"Depois de remover '{item_a_remover}': {convidados}")
else:
    print(f"'{item_a_remover}' não encontrado na lista.")
```

## 5. Operador `in`

O operador `in` é usado para verificar se um item específico está presente em uma lista. Ele retorna `True` se o item for encontrado e `False` caso contrário.

```python
cores = ["vermelho", "verde", "azul", "amarelo"]

tem_verde = "verde" in cores
print(f"'verde' está na lista? {tem_verde}")

tem_roxo = "roxo" in cores
print(f"'roxo' está na lista? {tem_roxo}")

# O operador not in também pode ser usado
nao_tem_preto = "preto" not in cores
print(f"'preto' NÃO está na lista? {nao_tem_preto}")
```

## 6. Contagem de itens

Para contar quantas vezes um item específico aparece em uma lista, usamos o método `.count()`.
Sintaxe: `lista.count(item_a_contar)`

```python
numeros_repetidos = [5, 10, 25, 50, 100, 500]

contagem_de_5 = numeros_repetidos.count(5)
print(f"O número 5 aparece {contagem_de_5} vezes.")

contagem_de_10 = numeros_repetidos.count(10)
print(f"O número 10 aparece {contagem_de_10} vezes.")

contagem_de_50 = numeros_repetidos.count(50) # Item não presente
print(f"O número 50 aparece {contagem_de_50} vezes.")

letras = ['a', 'b', 'a', 'c', 'a', 'd', 'a']
contagem_de_a = letras.count('a')
print(f"A letra 'a' aparece {contagem_de_a} vezes.")
```

## 7. Maior e menor

Para encontrar o maior e o menor item em uma lista, podemos usar as funções `max()` e `min()`, respectivamente.
Essas funções funcionam bem com listas de números e também com listas de strings (onde a ordenação é lexicográfica, ou seja, alfabética).

**Com números:**

```python
saldo_final = [880, 920, 750, 99, 810, 75]

maior_pontuacao = max(saldo_final)
print(f"A maior pontuação foi: {maior_pontuacao}")

menor_pontuacao = min(saldo_final)
print(f"A menor pontuação foi: {menor_pontuacao}")
```

**Com strings:**
A comparação é feita caractere por caractere, seguindo a ordem do alfabeto (e a tabela Unicode para outros caracteres). Letras maiúsculas vêm antes das minúsculas.

```python
nomes = ["Carlos", "ana", "Bruno", "Daniela", "Beatriz"]

maior_nome = max(nomes) # Lexicograficamente, "ana" é o maior porque 'a' > 'C', 'B', 'D'
print(f"O 'maior' nome (lexicograficamente) é: {maior_nome}")

menor_nome = min(nomes) # Lexicograficamente, "Beatriz" é o menor
print(f"O 'menor' nome (lexicograficamente) é: {menor_nome}")

# Para uma comparação mais intuitiva de strings, geralmente convertemos para um mesmo caso (maiúsculo ou minúsculo)
# mas max() e min() usarão a ordem padrão.
```

**Importante:** Não se pode usar `max()` ou `min()` em listas com tipos de dados mistos que não podem ser comparados entre si (por exemplo, uma lista contendo números e strings ao mesmo tempo).

## 8. Erros comuns com Listas

Ao trabalhar com listas, alguns erros são bastante comuns, especialmente para iniciantes. Conhecê-los pode ajudar a evitá-los ou a corrigi-los mais rapidamente.

### 8.1. `IndexError: list index out of range`
Este é provavelmente o erro mais comum. Ele ocorre quando você tenta acessar um índice que não existe na lista. Lembre-se que os índices de uma lista de tamanho `N` vão de `0` até `N-1`.

```python
minha_lista_idx = [10, 20, 30]
print(f"A lista é: {minha_lista_idx}")
print(f"O tamanho da lista é: {len(minha_lista_idx)}")
# print(minha_lista_idx[3]) # Descomente para ver o erro. O índice 3 não existe (índices válidos: 0, 1, 2)
# print(minha_lista_idx[-4]) # Descomente para ver o erro. O índice -4 não existe (índices negativos válidos: -1, -2, -3)
```

**Como evitar:**
- Sempre verifique o tamanho da lista com `len()` antes de acessar um índice calculado.
- Tenha cuidado com os limites em laços (loops) que acessam itens por índice.

### 8.2. `ValueError: <item> is not in list`
Este erro ocorre quando você usa o método `.remove(item)` para tentar remover um item que não está presente na lista.

```python
# Exemplo de ValueError
lista_valores = ["A", "B", "C"]
print(f"A lista é: {lista_valores}")
# lista_valores.remove("D") # Descomente para ver o erro. "D" não está na lista.
```

**Como evitar:**
- Use o operador `in` para verificar se o item existe na lista antes de tentar removê-lo:

```python
lista_valores_segura = ["A", "B", "C"]
item_para_remover = "D"

if item_para_remover in lista_valores_segura:
    lista_valores_segura.remove(item_para_remover)
else:
    print(f"O item '{item_para_remover}' não foi encontrado para remoção.")
print(f"Lista após tentativa de remoção segura: {lista_valores_segura}")
```

### 8.3. Esquecer que os índices começam em 0
Muitos iniciantes esquecem que o primeiro elemento de uma lista está no índice `0`, não no `1`. Isso pode levar a erros de lógica ou `IndexError`.

Prestar atenção a esses pontos pode economizar muito tempo de depuração!
