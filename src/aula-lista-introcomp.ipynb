## Dinmica da Caixa

```python
import random
import time

# 1. Boas-vindas e Configurao Inicial
# ----------------------------------------
print("---  Bem-vindo(a)  Roleta (Verso com Listas)! ---")
time.sleep(1)

# Esta  a LISTA de valores que sero sorteados para formar o saldo do jogador.
valores_pontos = [5, 10, 25, 50, 100, 500]

num_jogadores = int(input("\nQuantos jogadores vo participar? "))

# 2. Coleta de Apostas usando Listas
# ---------------------------------------------
# Vamos usar vrias LISTAS para guardar as informaes. Cada lista guarda
# um tipo de dado, e a POSIO (ndice) conecta as informaes do mesmo jogador.

# Exemplo para 2 jogadores (Ana no ndice 0, Beto no ndice 1):
# nomes_jogadores =   ["Ana", "Beto"]
# numeros_apostados = [10,    25]
# saldos_iniciais =   [150,   85]

nomes_jogadores = []
numeros_apostados = []
saldos_iniciais = []
total_apostado = 0

print("\n--- Hora de fazer as apostas! ---")
# O `for` vai repetir o bloco de cdigo para cada jogador.
for i in range(num_jogadores):
    print(f"\n--- Vez do Jogador {i+1} ---")

    nome = input(f"Digite o nome do jogador {i+1}: ")

    # Calcula o saldo inicial sorteando 3 fichas e somando.
    saldo = 0
    for _ in range(3):
        ficha_sorteada = random.choice(valores_pontos)
        saldo += ficha_sorteada

    print(f" {nome}, voc sorteou 3 fichas e seu saldo  de {saldo} pontos.")

    numero = int(input(f"Em qual nmero voc quer apostar (de 0 a 36)? "))

    # O jogador aposta todo o seu saldo (All-in).
    print(f"Voc apostou todos os seus {saldo} pontos no nmero {numero}.")

    # Adiciona as informaes do jogador nas nossas listas.
    # O mtodo `.append()` adiciona um item ao FINAL da lista.
    # A ordem  importante para manter os dados alinhados!
    nomes_jogadores.append(nome)
    numeros_apostados.append(numero)
    saldos_iniciais.append(saldo)

    # Soma a aposta ao total. Como  'all-in', a aposta  o saldo.
    total_apostado += saldo

# 3. Girando a Roleta
# -------------------
print("\n----------------------------------------------------")
print("\nTodas as apostas foram feitas!")
print(f"O valor total na mesa  de {total_apostado} pontos.")
print("\n A roleta est girando...")
time.sleep(3) # Pausa maior para mais suspense.

numero_vencedor = random.randint(0, 36)
print(f"\n O nmero sorteado : {numero_vencedor}!")
time.sleep(2)

# 4. Verificando quem ganhou usando os NDICES da lista
# ------------------------------------------------------
print("\n---  Resultado Final  ---")

# Vamos criar uma LISTA para guardar os NDICES dos jogadores que ganharam.
indices_vencedores = []
# `range(num_jogadores)` vai gerar os ndices: 0, 1, 2...
for i in range(num_jogadores):
    # Verificamos se o nmero na lista `numeros_apostados` na posio `i`  o vencedor.
    if numeros_apostados[i] == numero_vencedor:
        # Se for, guardamos o NDICE `i` na nossa lista de vencedores.
        indices_vencedores.append(i)

# 5. Calculando e Mostrando os Resultados
# -----------------------------------------
# `len()` retorna o tamanho da lista.
if len(indices_vencedores) > 0:
    premio_por_vencedor = total_apostado / len(indices_vencedores)
    print(f"\nTemos {len(indices_vencedores)} vencedor(es)! O prmio de {total_apostado} pontos ser dividido.")

    # Passamos pela lista de NDICES dos vencedores.
    for i in indices_vencedores:
        # Usamos o ndice `i` para pegar o nome do vencedor na lista `nomes_jogadores`.
        print(f"-> Parabns, {nomes_jogadores[i]}! Voc ganhou {premio_por_vencedor:.2f} pontos.")
else:
    print("\n Ningum acertou o nmero! A casa (o computador) ganha.")

# 6. Balano Final de Pontos
# ----------------------------
print("\n---  Balano Final dos Jogadores ---")
saldos_finais = [] # Criamos uma nova lista para guardar os saldos finais
for i in range(num_jogadores):
    # Pegamos as informaes do jogador na posio `i` de cada lista.
    nome_jogador = nomes_jogadores[i]
    saldo_inicial_jogador = saldos_iniciais[i]

    # Como a aposta foi 'all-in', o saldo inicial vira 0, a no ser que ganhe.
    saldo_final = 0

    # Se o ndice `i` estiver na lista de vencedores, o jogador recebe o prmio.
    if i in indices_vencedores:
        saldo_final = premio_por_vencedor

    saldos_finais.append(saldo_final) # Adicionamos o saldo final  lista

    print(f"\nJogador: {nome_jogador}")
    print(f"  - Saldo Inicial: {saldo_inicial_jogador} pontos")
    print(f"  - Saldo Final: {saldo_final} pontos")

# 7. Exibindo as Listas Finais para Demonstrao
# -----------------------------------------------
print("\n----------------------------------------------------")
print("Contedo final das listas usadas no jogo:")
print(f"  -> nomes_jogadores:   {nomes_jogadores}")
print(f"  -> numeros_apostados: {numeros_apostados}")
print(f"  -> saldos_iniciais:   {saldos_iniciais}")
print(f"  -> saldos_finais:     {saldos_finais}")
print(f"  -> indices_vencedores:  {indices_vencedores}")

print("\n--- Fim do Jogo ---")
```

## 2. Introduo a Listas

Uma **lista** em Python  uma coleo ordenada e mutvel de itens. Isso significa que os itens tm uma ordem definida (e essa ordem no mudar a menos que voc a altere) e voc pode mudar, adicionar ou remover itens depois que a lista foi criada.

Os itens em uma lista, tambm chamados de **elementos**, podem ser de qualquer tipo de dado, e uma nica lista pode at conter itens de tipos diferentes (embora seja mais comum ter listas com itens do mesmo tipo).

Cada item em uma lista possui um **ndice**, que  um nmero indicando sua posio. A contagem dos ndices comea em `0`. Ou seja, o primeiro item tem ndice `0`, o segundo item tem ndice `1`, e assim por diante.

**Sintaxe para criar uma lista:**


> < nome da lista > = [ elemento1, elemento2, elemento3, ...]



```python
valores_pontos = [5, 10, 25]
print("Valores dos pontos:", valores_pontos)
print("-" * 20)
```

Para acessar um item especfico da lista, usamos seu ndice entre colchetes:

```python
print("Acessando Itens da Lista:\n")
primeiro_valor = valores_pontos[0]
terceiro_valor = valores_pontos[2]
print("Primeiro valor:", primeiro_valor)
print("Terceiro valor:", terceiro_valor)
print("-" * 20)
```

Como as listas so mutveis, podemos alterar um item existente:

```python
print(f"Lista antes da alterao: {valores_pontos}\n")
valores_pontos[2] = 50
print(f"Lista depois da alterao: {valores_pontos}")
```

###3. Adicionar e Remover da lista

Listas so dinmicas, o que significa que podemos adicionar e remover itens delas facilmente.

### 3.1. Adicionando itens

#### `append()`: Adiciona um item ao final da lista.

```python
nomes_jogadores = ["Luiz", "Artur"]
print(f"Lista inicial: {nomes_jogadores}")

print("\n## Adicionando Itens  Lista\n")
print("Valores antes de adicionar:", nomes_jogadores)
nomes_jogadores.append("Rafael")
nomes_jogadores.append("Guilherme")
print("Valores depois de adicionar novos nomes:", nomes_jogadores)
print("-" * 20)
```

#### `insert()`: Adiciona um item em uma posio especfica.
A sintaxe  `lista.insert(indice, item)`.

```python
nomes_jogadores = ["Luiz", "Artur"]
print(f"Lista inicial: {nomes_jogadores}")

nomes_jogadores.insert(1, "Guilherme") # Insere "Guilherme" no ndice 1
print(f"Depois de insert(1, 'Guilherme'): {nomes_jogadores}")

nomes_jogadores.insert(2, "Rafael") # Insere "Artur" no incio
print(f"Depois de insert(2, 'Artur'): {nomes_jogadores}")
```

### 3.1. Removendo itens

#### `pop()`: Remove e retorna o item de uma posio especfica. Se nenhum ndice for fornecido, remove e retorna o ltimo item.

```python
valores_pontos = ["5", "10", "25", "50", "100", "500"]

print("## Removendo Itens da Lista")
print(f"\nValores antes de remover: {valores_pontos}\n")
# Removendo o valor 10, 25, 500
valores_pontos.remove("10")
valores_pontos.remove("25")
valores_pontos.remove("500")

print(f"Valores depois de remover o 10, 25 e 500: {valores_pontos}\n")

valores_pontos.pop(2)
print(f"Valores depois de remover o item no ndice 2: {valores_pontos}\n")
print("-" * 20)
```

#### `remove()`: Remove a primeira ocorrncia de um item com um valor especfico.
Se o item no existir na lista, um erro (`ValueError`) ser gerado.

```python
convidados = ["Alice", "Bob", "Charlie", "David", "Bob"]
print(f"Convidados iniciais: {convidados}")

convidados.remove("Bob") # Remove a primeira ocorrncia de "Bob"
print(f"Depois de remove('Bob'): {convidados}")

convidados.remove("David")
print(f"Depois de remove('David'): {convidados}")

# Tentando remover um item que no existe (isso geraria um erro, ento vamos verificar antes)
item_a_remover = "Eve"
if item_a_remover in convidados:
    convidados.remove(item_a_remover)
    print(f"Depois de remover '{item_a_remover}': {convidados}")
else:
    print(f"'{item_a_remover}' no encontrado na lista.")
```

## 5. Operador `in`

O operador `in`  usado para verificar se um item especfico est presente em uma lista. Ele retorna `True` se o item for encontrado e `False` caso contrrio.

```python
cores = ["vermelho", "verde", "azul", "amarelo"]

tem_verde = "verde" in cores
print(f"'verde' est na lista? {tem_verde}")

tem_roxo = "roxo" in cores
print(f"'roxo' est na lista? {tem_roxo}")

# O operador not in tambm pode ser usado
nao_tem_preto = "preto" not in cores
print(f"'preto' NO est na lista? {nao_tem_preto}")
```

## 6. Contagem de itens

Para contar quantas vezes um item especfico aparece em uma lista, usamos o mtodo `.count()`.
Sintaxe: `lista.count(item_a_contar)`

```python
numeros_repetidos = [5, 10, 25, 50, 100, 500]

contagem_de_5 = numeros_repetidos.count(5)
print(f"O nmero 5 aparece {contagem_de_5} vezes.")

contagem_de_10 = numeros_repetidos.count(10)
print(f"O nmero 10 aparece {contagem_de_10} vezes.")

contagem_de_50 = numeros_repetidos.count(50) # Item no presente
print(f"O nmero 50 aparece {contagem_de_50} vezes.")

letras = ['a', 'b', 'a', 'c', 'a', 'd', 'a']
contagem_de_a = letras.count('a')
print(f"A letra 'a' aparece {contagem_de_a} vezes.")
```

## 7. Maior e menor

Para encontrar o maior e o menor item em uma lista, podemos usar as funes `max()` e `min()`, respectivamente.
Essas funes funcionam bem com listas de nmeros e tambm com listas de strings (onde a ordenao  lexicogrfica, ou seja, alfabtica).

**Com nmeros:**

```python
saldo_final = [880, 920, 750, 99, 810, 75]

maior_pontuacao = max(saldo_final)
print(f"A maior pontuao foi: {maior_pontuacao}")

menor_pontuacao = min(saldo_final)
print(f"A menor pontuao foi: {menor_pontuacao}")
```

**Com strings:**
A comparao  feita caractere por caractere, seguindo a ordem do alfabeto (e a tabela Unicode para outros caracteres). Letras maisculas vm antes das minsculas.

```python
nomes = ["Carlos", "ana", "Bruno", "Daniela", "Beatriz"]

maior_nome = max(nomes) # Lexicograficamente, "ana"  o maior porque 'a' > 'C', 'B', 'D'
print(f"O 'maior' nome (lexicograficamente) : {maior_nome}")

menor_nome = min(nomes) # Lexicograficamente, "Beatriz"  o menor
print(f"O 'menor' nome (lexicograficamente) : {menor_nome}")

# Para uma comparao mais intuitiva de strings, geralmente convertemos para um mesmo caso (maisculo ou minsculo)
# mas max() e min() usaro a ordem padro.
```

**Importante:** No se pode usar `max()` ou `min()` em listas com tipos de dados mistos que no podem ser comparados entre si (por exemplo, uma lista contendo nmeros e strings ao mesmo tempo).

## 8. Erros comuns com Listas

Ao trabalhar com listas, alguns erros so bastante comuns, especialmente para iniciantes. Conhec-los pode ajudar a evit-los ou a corrigi-los mais rapidamente.

### 8.1. `IndexError: list index out of range`
Este  provavelmente o erro mais comum. Ele ocorre quando voc tenta acessar um ndice que no existe na lista. Lembre-se que os ndices de uma lista de tamanho `N` vo de `0` at `N-1`.

```python
minha_lista_idx = [10, 20, 30]
print(f"A lista : {minha_lista_idx}")
print(f"O tamanho da lista : {len(minha_lista_idx)}")
# print(minha_lista_idx[3]) # Descomente para ver o erro. O ndice 3 no existe (ndices vlidos: 0, 1, 2)
# print(minha_lista_idx[-4]) # Descomente para ver o erro. O ndice -4 no existe (ndices negativos vlidos: -1, -2, -3)
```

**Como evitar:**
- Sempre verifique o tamanho da lista com `len()` antes de acessar um ndice calculado.
- Tenha cuidado com os limites em laos (loops) que acessam itens por ndice.

### 8.2. `ValueError: <item> is not in list`
Este erro ocorre quando voc usa o mtodo `.remove(item)` para tentar remover um item que no est presente na lista.

```python
# Exemplo de ValueError
lista_valores = ["A", "B", "C"]
print(f"A lista : {lista_valores}")
# lista_valores.remove("D") # Descomente para ver o erro. "D" no est na lista.
```

**Como evitar:**
- Use o operador `in` para verificar se o item existe na lista antes de tentar remov-lo:

```python
lista_valores_segura = ["A", "B", "C"]
item_para_remover = "D"

if item_para_remover in lista_valores_segura:
    lista_valores_segura.remove(item_para_remover)
else:
    print(f"O item '{item_para_remover}' no foi encontrado para remoo.")
print(f"Lista aps tentativa de remoo segura: {lista_valores_segura}")
```

### 8.3. Esquecer que os ndices comeam em 0
Muitos iniciantes esquecem que o primeiro elemento de uma lista est no ndice `0`, no no `1`. Isso pode levar a erros de lgica ou `IndexError`.

Prestar ateno a esses pontos pode economizar muito tempo de depurao!
